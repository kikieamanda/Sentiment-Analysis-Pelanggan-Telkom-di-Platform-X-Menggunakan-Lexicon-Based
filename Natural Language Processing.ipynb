# Preprocessing

```python
columns_to_drop = [
    'favorite_count', 'id_str', 'image_url', 'in_reply_to_screen_name', 'lang',
    'quote_count', 'reply_count', 'retweet_count', 'user_id_str', 'tweet_url'
]

# Menghapus kolom-kolom tersebut dari DataFrame
Telkom = telkom.drop(columns=columns_to_drop)
Telkom
```

```python
# Daftar kata/frasa yang ingin dihapus
phrases_to_remove = [
    'd semua benar',
    'D Semua benar',
    '#QuizKamisDigimin',
    'cek dm',
    'Ibu Kota Nusantara (IKN) memiliki jaringan komunikasi dan konektivitas andal yang disediakan oleh salah satu perusahaan telekomunikasi'
]

def contains_phrases(text):
    for phrase in phrases_to_remove:
        if phrase in text:
            return True
    return False

# Menghapus baris yang mengandung kata/frasa di kolom 'teks'
Telkom_cleaned = Telkom[~Telkom['Text'].apply(lambda x: contains_phrases(str(x)))]
Telkom_cleaned
```

```python
# Daftar username yang ingin dihapus
usernames_to_remove = ['TelkomIndonesia', 'Indibiz_ID', 'IndiHomeCare', 'Telkomsel', 'TENESA_TELKOM', 'TelkomSolution', 'smbtelkom']

# Menghapus baris yang mengandung username tertentu di kolom 'username'
Telkom_Cleaned = Telkom_cleaned[~Telkom_cleaned['username'].isin(usernames_to_remove)]
Telkom_Cleaned
```

```python
import re
import string

# Fungsi untuk membersihkan teks
def clean_text(text):
    # Mengubah teks menjadi huruf kecil
    text = text.lower()
    # Menghapus link
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Menghapus tag, hashtag, dan angka
    text = re.sub(r'#\w+|\@\w+|\d+', '', text)
    # Menghapus tanda baca
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Menghapus whitespace
    text = ' '.join(text.split())
    return text

# Mengecek duplikat pada kolom 'Text' dan menghapus baris yang duplikat
telkom_cleaned = Telkom_Cleaned.drop_duplicates(subset=['Text'], keep='first')

# Mengubah teks di kolom 'Text' menjadi bersih
telkom_cleaned['Text'] = telkom_cleaned['Text'].apply(clean_text)

# Menampilkan jumlah data setelah menghapus duplikat dan membersihkan teks
print("Jumlah data setelah menghapus duplikat dan membersihkan teks:", len(telkom_cleaned))

# Menyimpan DataFrame yang telah dibersihkan ke file CSV baru
telkom_cleaned.to_csv("telkomaduan_cleaned.csv", index=False)

# Menampilkan beberapa data untuk memastikan
telkom_cleaned
```

```python
from nltk.tokenize import word_tokenize
import nltk

# Downloading necessary NLTK data
nltk.download('punkt')
nltk.download('punkt_tab')

# Fungsi untuk melakukan tokenisasi
def tokenize_text(text):
    return word_tokenize(text)

# Melakukan tokenisasi pada kolom 'Text'
telkom_cleaned['Tokens'] = telkom_cleaned['Text'].apply(tokenize_text)

# Menyimpan DataFrame yang telah dibersihkan dan ditokenisasi ke file CSV baru
telkom_cleaned.to_csv("telkomaduan_cleaned_tokenized.csv", index=False)

# Menampilkan beberapa data untuk memastikan
telkom_cleaned
```

```python
from nltk.corpus import stopwords
import nltk

nltk.download('stopwords')

# Fungsi untuk menghapus stopwords
def remove_stopwords(text):
    stop_words = set(stopwords.words('indonesian'))  # Gunakan 'english' jika menggunakan bahasa Inggris
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]
    return ' '.join(filtered_words)

# Menghapus stopwords dari kolom 'Text'
telkom_cleaned['Text'] = telkom_cleaned['Text'].apply(remove_stopwords)

# Menyimpan DataFrame yang telah dibersihkan ke file CSV baru
telkom_cleaned.to_csv("telkomaduan_cleaned_no_stopwords.csv", index=False)

# Menampilkan beberapa data untuk memastikan
telkom_cleaned
```

```python
import pandas as pd
telkom1 = pd.read_csv("telkomaduan_sentiment_score.csv")
telkom1
```

```python
def classify_sentiment(score):
    if score > 0:
        return '+'  # Positif
    elif score < 0:
        return '-'  # Negatif
    else:
        return '0'  # Netral

telkom1['Sentiment_Score'] = telkom1['Sentiment_Score'].apply(classify_sentiment)

# Menghitung jumlah sentimen positif, negatif, dan netral
sentiment_counts = telkom1['Sentiment_Score'].value_counts()

print("Jumlah Sentimen Positif (+):", sentiment_counts.get('+', 0))
print("Jumlah Sentimen Negatif (-):", sentiment_counts.get('-', 0))
print("Jumlah Sentimen Netral (0):", sentiment_counts.get('0', 0))
```

```python
import matplotlib.pyplot as plt

# Menghitung jumlah sentimen positif, negatif, dan netral
sentiment_counts = telkom1['Sentiment_Score'].value_counts()

# Membuat grafik batang (bar plot)
plt.figure(figsize=(8, 6))
sentiment_counts.plot(kind='bar', color=['green', 'grey', 'red'])
plt.title('Distribusi Sentimen')
plt.xlabel('Sentimen Publik')
plt.ylabel('Jumlah')
plt.xticks(ticks=[0, 1, 2], labels=['Netral', 'Positif', 'Negatif'], rotation=0)
plt.show()
```

```python
from sklearn.metrics import f1_score, classification_report

# Ekstrak true_labels dan predicted_labels
true_labels = telkom1["Label Asli"]
predicted_labels = telkom1["Sentiment_Score"]

true_labels = true_labels.astype(str)
predicted_labels = predicted_labels.astype(str)

f1_macro = f1_score(true_labels, predicted_labels, average='macro')  # Rata-rata antar kelas
f1_micro = f1_score(true_labels, predicted_labels, average='micro')  # Mengabaikan distribusi kelas

print("F1-Score (Macro):", f1_macro)
print("F1-Score (Micro):", f1_micro)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report

# Fungsi untuk evaluasi model sentimen
def evaluate_sentiment_model(true_labels, predicted_labels):
    # Precision
    precision = precision_score(true_labels, predicted_labels, average='weighted')
    # Recall
    recall = recall_score(true_labels, predicted_labels, average='weighted')
    # F1-Score
    f1 = f1_score(true_labels, predicted_labels, average='weighted')

    # Menampilkan laporan lengkap
    print("Classification Report:")
    print(classification_report(true_labels, predicted_labels))

    return precision, recall, f1

# Evaluasi model
precision, recall, f1 = evaluate_sentiment_model(true_labels, predicted_labels)

# Menampilkan hasil
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
```
